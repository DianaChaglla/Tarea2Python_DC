Se importan las librerías que seran de utilidad para la manipulación de los datos y la realización de los gráficos, `pandas` y `matplotlib` respectivamente.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Se importan los datos procesados directamente desde Github, la base de datos contiene información sobre los partidos de todas las ediciones de la Copa del Mundo de fútbol masculino.

```python
data_mundial = pd.read_csv("https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt", sep = "\t")
data_mundial
```

Se realiza una filtración con la información de los países que son de interés, para el estudio comparativo, en las ocasiones en las que fueron el equipo 1 y el equipo 2. Para este caso en particular los países seleccionados, arbitrariamente, fueron Argentina y Ecuador. 

```python
#Creacion de dataframe filter Argentina
arg_equipo1 = data_mundial[data_mundial["equipo_1"] == "Argentina"]
arg_equipo2 = data_mundial[data_mundial["equipo_2"] == "Argentina"]
argentina = pd.concat([arg_equipo1, arg_equipo2])
argentina.index = ['ARG'] * len(argentina)
argentina
```

```python
#Creacion de dataframe filter Ecuador
ec_equipo1 = data_mundial[data_mundial["equipo_1"] == "Ecuador"]
ec_equipo2 = data_mundial[data_mundial["equipo_2"] == "Ecuador"]
ecuador = pd.concat([ec_equipo1, ec_equipo2])
ecuador.index = ['EC'] * len(ecuador)
ecuador
```

Una vez creados los dataframes por cada país seleccionado se usa la función `groupby()` para calcular la cantidad total de partidos jugados en cada mundial en el que el país clasifico y la función `rename()` para renombrar la columna utilizada como referencia para el conteo y la gráfica porsterior. 

```python
arg_tp = argentina.groupby('anio')["equipo_1_final"].count().to_frame()
ec_tp = ecuador.groupby('anio')["equipo_1_final"].count().to_frame()
arg_tp.rename(
    columns={"equipo_1_final": "Argentina"},
    inplace=True,
)
ec_tp.rename(
    columns={"equipo_1_final": "Ecuador"},
    inplace=True,
)
display(arg_tp)
display(ec_tp)
```

```python
#Creamos una figura (imaginate un lienzo en blanco)
fig = plt.figure()
fig.suptitle("Cantidad de partidos jugados")
#Agregamos un eje (un sistema de coordenadas donde graficaremos)
eje = fig.add_subplot(1, 1, 1)
#Ahora graficamos los datos de Chile y Ecuador en el mismo eje
arg_tp.plot(ax = eje)
ec_tp.plot(ax = eje)
#Ahora etiquetamos los ejes
plt.xlabel('Año')
plt.ylabel('Cantidad de partidos')
eje.legend(loc='upper left', borderpad=1.5, labelspacing=1.5)
plt.savefig('graph1.jpg') #funcion para guardar el gráfico
```

Como se puede ver en el gráfico *Cantidad de partidos jugados* Argentina ha jugado mayor cantidad de partidos en comparación con Ecuador, debido a que ha participado en 17 de las 21 ediciones mundialistas hasta la fecha corte de los datos en el año 2019, Ecuador en cambio unicamente ha participado en 3 ediciones. Ecuador ha jugado un máximo de 4 partidos que fueron disputados en el mundial de Alemania 2006, por otro lado la máxima cantidad de partidos jugados por Argentinda es 7 en los mundiales de 1978, 1986, 1990 y 2014, conociendo que esa es la cantidad máxima de partidos que se pueden disputar en todas las fases del mundial se puede inferir que en esos años Argentina al menos llego a ser finalista. 

Ahora que ya se conoce de manera general que la participación de la selección Argentina ha sido superior a la Ecuatoriana se quiere calcular la cantidad de goles marcados en cada mundial jugado. Para esa finalidad se usa nuevamente la función `groupby()` acompañada de una sumataria que nos permite calcular los goles realizados tanto en las ocasiones en las que las selecciones fueron el *equipo 1* y *equipo 2*. Es importante notar que para poder hacer uso del gráfico como una imagen fuera del ambiente se usa la función `plt.savefig()` para guardarla en nuestro computador.

```python
arg1_y = arg_equipo1.groupby('anio')[["equipo_1_final", "equipo_2_final"]].sum()
ec1_y = ec_equipo1.groupby('anio')[["equipo_1_final", "equipo_2_final"]].sum()
arg2_y = arg_equipo2.groupby('anio')[["equipo_1_final", "equipo_2_final"]].sum()
ec2_y = ec_equipo2.groupby('anio')[["equipo_1_final", "equipo_2_final"]].sum()
```

```python
#renaming columns
arg1_y.rename(
    columns={"equipo_1_final": "Goles Argentina", "equipo_2_final": "Goles rival"},
    inplace=True,
)
ec1_y.rename(
    columns={"equipo_1_final": "Goles Ecuador", "equipo_2_final": "Goles rival"},
    inplace=True,
)
arg2_y.rename(
    columns={"equipo_1_final": "Goles rival", "equipo_2_final": "Goles Argentina"},
    inplace=True,
)
ec2_y.rename(
    columns={"equipo_1_final": "Goles rival", "equipo_2_final": "Goles Ecuador"},
    inplace=True,
)
display(arg1_y)
display(arg2_y)
display(ec1_y)
display(ec2_y)
```

```python
#Inicializamos una figura
fig = plt.figure()
fig.suptitle("Cantidad de goles por Mundial participado")
#plt.title('Cantidad de goles por partidos jugados en el mundial')
eje = fig.add_subplot(2, 2, 1)
arg1_y.plot(kind = "bar", ax = eje)
eje2 = fig.add_subplot(2, 2, 2, sharey = eje)
ec1_y.plot(kind = "bar", ax = eje2)
eje3 = fig.add_subplot(2, 2, 3)
arg2_y.plot(kind = "bar", ax = eje3)
eje4 = fig.add_subplot(2, 2, 4, sharey = eje)
ec2_y.plot(kind = "bar", ax = eje4)
#Agregamos etiquetas a los ejes
eje.set_xlabel('Año')
eje2.set_xlabel('Año')
eje3.set_xlabel('Año')
eje4.set_xlabel('Año')
eje.set_ylabel('Goles')
eje2.set_ylabel('Goles')
eje3.set_ylabel('Goles')
eje4.set_ylabel('Goles')
#subplot title
eje.title.set_text('Argentina Equipo 1')
eje2.title.set_text('Ecuador Equipo 1')
eje3.title.set_text('Argentina Equipo 2')
eje4.title.set_text('Ecuador Equipo 2')
#save
plt.savefig('graph2.jpg')
```

En el gráfico de barras *Cantidad de goles por mundial participado* se observan 4 subgráficos con el total de goles marcados por la selecciones escogigas comparadas con el total de los goles que marcaron los equipos adversarios. En el caso de Argentina se puede observar que cuando fueron denominados equipo 1 la cantidad de goles es igual o superior en casi todos los mundiales excepto en el de 2018, mientras que cuando es denominado el equipo 2 por lo general la cantidad de goles es inferior teniendo en cuenta también que se lograron 3 empates. 

Por el otro lado Ecuador cuando fue denominado equipo 1 tenemos 2 empates y una ganancia, mientras que cuando fue el equipo 2 tenemos 1 victoria, 1 empate y 1 derrota.
